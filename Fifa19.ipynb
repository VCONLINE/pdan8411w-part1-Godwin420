The key assumptions of linear regression and why they are crucial for ensuring the robustness and reliability of the analysis:

Linearity: This assumption posits that there exists a linear relationship between the independent variables and the dependent variable. In other words, the change in the dependent variable is directly proportional to changes in the independent variables. Linear regression models assume that the relationship can be adequately captured by a straight line. Deviations from linearity can lead to biased estimates and inaccurate predictions. (Kaggle, 7 April 2024).

Independence: Independence assumes that the observations or data points used in the regression analysis are not influenced by each other. Each observation should be independent of all others. Violations of independence, such as autocorrelation in time-series data or clustering effects in clustered data, can lead to biased standard errors and inflated significance levels. (Kaggle, 7 April 2024).

Homoscedasticity: Homoscedasticity refers to the assumption that the variance of the residuals (the differences between observed and predicted values) is constant across all levels of the independent variables. In simpler terms, it implies that the spread of the residuals should be uniform across the range of predicted values. Heteroscedasticity, where the spread of residuals varies across different levels of the independent variable, can result in inefficient parameter estimates and inaccurate confidence intervals. (Kaggle, 7 April 2024).

Normality of Residuals: This assumption states that the residuals of the regression model should be normally distributed. Normality ensures that the statistical tests used for inference, such as hypothesis testing and confidence intervals, are valid. Departures from normality can affect the accuracy of parameter estimates and lead to incorrect conclusions about the significance of predictors. (Kaggle, 7 April 2024).

Ensuring that these assumptions hold true is essential for the validity and reliability of the linear regression analysis. Failure to meet these assumptions can result in biased estimates, unreliable predictions, and erroneous conclusions. (Kaggle, 7 April 2024).

## What is linear regression

Linear regression is a statistical model used to estimate the relationship between a dependent variable and one or more independent variables. Wikipedia. (7 April 2024). It is a technique used to model the relationships between observed variables. Brilliant Math & Science Wiki. (7 April 2024). The simplest form of linear regression involves two variables: a dependent variable (y) and an independent variable (x). Britannica. (2024). 

The main idea behind simple linear regression is to “fit” the observations of two variables into a linear relationship between them. Brilliant Math & Science Wiki. (7 April 2024). This is achieved by determining a line that best represents the general trend of a data set. Britannica. (2024). Regression models describe the relationship between variables by fitting a line to the observed data. Scribbr. (2023). Linear regression models use a straight line, while logistic and nonlinear regression models use a curved line. Scribbr. (2023)

For example, if you are a social researcher interested in the relationship between income and happiness, you might survey 500 people whose incomes range from 15k to 75k and ask them to rank their happiness on a scale from 1 to 10. Your independent variable (income) and dependent variable (happiness) are both quantitative, so you can do a regression analysis to see if there is a linear relationship between them. Scribbr. (2023). 

Linear regression makes certain assumptions about the data: Scribbr. (2023).

Homogeneity of variance (homoscedasticity): the size of the error in our prediction doesn’t change significantly across the values of the independent variable.
Independence of observations: the observations in the dataset were collected using statistically valid sampling methods, and there are no hidden relationships among observations.

Normality: The data follows a normal distribution.
The relationship between the independent and dependent variable is linear: the line of best fit through the data points is a straight line (rather than a curve or some sort of grouping factor).
If your data do not meet the assumptions of homoscedasticity or normality, you may be able to use a nonparametric test instead, such as the Spearman rank test.Scribbr. (2023)



The equation for a simple linear regression is typically written as:
$y=a+bx$
where:

$y$ is the dependent variable (the variable we are trying to predict or estimate)
$x$ is the independent variable (the variable we are using to make the prediction)
$a$ is the y-intercept (the value of y when x=0)
$b$ is the slope of the regression line (the change in y for a one-unit change in x). BYJU’S. (7 April 2024).

In the context of multiple linear regression, where we have more than one independent variable, the equation becomes:
$y=a+b1​x1​+b2​x2​+...+bn​xn​$
where:

$x1​,x2​,...,xn$​ are the independent variables
$b1​,b2​,...,bn$​ are the coefficients of the independent variables (the change in y for a one-unit change in the respective independent variable)
a is the y-intercept (the value of y when all x=0)

The coefficients $b1​,b2​,...,bn​$ and the y-intercept $a$ are typically estimated using a method called least squares. Statistics By Jim. (7 April 2024). This method minimizes the sum of the squared differences between the observed dependent variable in the dataset and those predicted by the linear function. Statistics By Jim. (7 April 2024).

```python
#removing spaces from columns
df.columns = [c.replace(' ', '') for c in df.columns]
df.columns
```

```python
#Creating Dateframe with playars BELOW 85 overall
df = df.query("Overall<=85")
df.head()
```

```python
#remove data not needed or wanted
df.drop(["Unnamed:0", "GKDiving", "GKHandling", "GKKicking", "GKPositioning", "GKReflexes",
       "ReleaseClause", "ID", "Name",  "Photo", "Nationality", "Flag",
        "Club", "ClubLogo", "Special",
       "WorkRate", "BodyType", "RealFace", "Position", "JerseyNumber",
       "Joined", "LoanedFrom", "ContractValidUntil", "Height", "Weight", "LS",
       "ST", "RS", "LW", "LF", "CF", "RF", "RW", "LAM", "CAM", "RAM", "LM",
       "LCM", "CM", "RCM", "RM", "LWB", "LDM", "CDM", "RDM", "RWB", "LB",
       "LCB", "CB", "RCB", "RB",],axis=1,inplace=True)
df.head()
```

```python
df.tail()
```

```python
#Getting rid of currency sign
df['Wage'] = df['Wage'].str[1:]
df['Value'] = df['Value'].str[1:]

df.head()
```

```python
df.dtypes
```

```python
#Converting the value column to float, ie, if value has k multiply by 1000 and M 1e6
def value_and_wage_conversion(Value):
    if isinstance(Value,str):
        out = Value.replace('€', '')
        if 'M' in out:
            out = float(out.replace('M', ''))*1000000
        elif 'K' in out:
            out = float(out.replace('K', ''))*1000
        return float(out)
    
df['Value'] = df['Value'].apply(value_and_wage_conversion)
df['Wage'] = df['Wage'].apply(value_and_wage_conversion)
df.head()
```

```python
# Select the features and the target
X = df[['Value','Age','WeakFoot','Potential','Wage','Crossing', 'Finishing', 'HeadingAccuracy', "SkillMoves","InternationalReputation",
      'ShortPassing', 'Volleys', 'Dribbling', 'Curve', 'FKAccuracy',
       'LongPassing', 'BallControl', 'Acceleration', 'SprintSpeed', 'Agility',
       'Reactions', 'Balance', 'ShotPower', 'Jumping', 'Stamina', 'Strength',
       'LongShots', 'Aggression', 'Interceptions', 'Positioning', 'Vision',
       'Penalties', 'Composure', 'Marking', 'StandingTackle', 'SlidingTackle']]



y = df['Overall']
```

```python
#Scaling data to ensure all features have the same scale
scaler = StandardScaler()
scaler.fit_transform(X)
```

```python
# Assuming that 'X' is your feature set and 'y' is the target variable
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# Define the cross-validation method
cv = RepeatedKFold(n_splits=10, n_repeats=6)

# Define the range of alpha values to try
alphas = np.arange(0.0001, 0.001, 0.01)

# Create a Ridge Regression model
ridge = RidgeCV(alphas=alphas, cv=cv)
```

```python
# Train the model
ridge.fit(X_train, y_train)

# The alpha_ attribute holds the best alpha value found during cross-validation
print(f'Best alpha: {ridge.alpha_}')
```

```python
# Now you can use the trained model to make predictions on your test set
y_pred = ridge.predict(X_test)
```

```python
# I want to see the actual value vs the predicted value
for actual, predicted in zip(y_test, y_pred):
     print(f'Actual: {actual}, Predicted: {predicted}')
```

```python
# Create a color array, which can be based on any metric. Here we use the absolute error.
colors = np.abs(y_test - y_pred)

plt.scatter(y_test, y_pred, c=colors, cmap='viridis')
plt.colorbar(label='Error')  # Add a colorbar
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual vs Predicted')
plt.show()

```

```python
# Assuming that 'y_test' are the actual values and 'y_pred' are the predicted values

# Calculate R-squared
r2 = r2_score(y_test, y_pred)
print(f'R-squared: {r2}')

# Calculate Mean Squared Error
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error: {mse}')

# Calculate Root Mean Squared Error
rmse = np.sqrt(mse)
print(f'Root Mean Squared Error: {rmse}')
print("Training set score: {:.2f}".format(ridge.score(X_train,y_train)))
print("Test set score: {:.2f}".format(ridge.score(X_test,y_test)))
```
